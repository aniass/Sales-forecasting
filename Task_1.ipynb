#### 4. Which of the products was the most popular? How many customers bought it at a discount?

First we connect the columns:

```python
data = pd.merge(df1, df3, on='order_id')
data.head()
```

We extract products, how many times a given product has occurred during individual orders:

```python
item1 = pd.DataFrame(data['item_id_1'].value_counts())
item1
```

```python
item2 = pd.DataFrame(data['item_id_2'].value_counts())
item2
```

```python
item3 = pd.DataFrame(data['item_id_3'].value_counts())
item3
```

```python
#we connect the obtained data into one column
dane = pd.concat([item1, item2, item3], axis=1).reset_index()
dane
```

```python
dane = dane.rename(columns={"index": "product"})
dane.head()
```

```python
#we check missing values
dane.isnull().sum()
```

```python
dane["item_id_3"] = dane["item_id_3"].fillna('0')
```

```python
dane.info()
```

```python
dane['item_id_3'] = dane['item_id_3'].astype(int)
```

We combine the occurrence of each product in individual orders together:

```python
dane['items'] = dane['item_id_1'] + dane['item_id_2'] + dane['item_id_3']
dane.head()
```

```python
dane.isnull().sum()
```

```python
dane.sort_values('items', ascending=False)
```

The most frequently ordered product was the reference number 4355722.

```python
data[data['item_id_3']== 4355722]
```
